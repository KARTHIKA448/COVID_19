### Comparisons of total cases with active, discharged and deaths

```python
plt.figure(figsize=(12, 5))
plt.subplot(1,3,1)
plt.hist(df['Active'], bins=20, edgecolor='black')
plt.title("Active VS Total Cases ",fontdict={'size':12,'color':'green'})
plt.xlabel("Total Cases")
plt.ylabel("Active")

plt.subplot(1,3,2)
plt.hist(df['Deaths'], bins=20, edgecolor='black')
plt.title("Discharged VS Total Cases",fontdict={'size':12,'color':'blue'})
plt.xlabel("Total Cases")
plt.ylabel("Discharged")

plt.subplot(1,3,3)
plt.hist(df['Total Cases'], bins=20, edgecolor='black')
plt.title("Deaths VS Total Cases",fontdict={'size':12,'color':'red'})
plt.xlabel("Total Cases")
plt.ylabel("Deaths")



plt.tight_layout()
plt.show()

# The first parameter represents the number of rows in the subplot grid.
# The second parameter represents the number of columns in the subplot grid.
# The third parameter represents the index or position of the subplot within the grid.
```

### RATIO ANALYSIS

```python
labels = ["Active","Deaths"]
sizes = [df["Active Ratio"].sum(),df["Death Ratio"].sum()]
```

```python
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=140)
plt.title("Covid_19 ratio analysis")
plt.show()
```

- out of total cases most are death cases

### Least number of discharged cases among each state

```python
discharged = np.array(df['Discharged'])
states = np.array(df['State'])

ddf = pd.DataFrame({'discharged':discharged,
                     'states':states}, index = states)

sortddf = ddf.discharged.sort_values(ascending=True).head()
print("least number of discharged cases among each state: ",sortddf)

sortddf.plot.bar(title="Least number of discharged cases among states",
               xlabel="States/UTs",
               ylabel="Number of discharges",
               rot=25,
               fontsize=10,
               legend=True)
plt.xticks(rotation=90)
plt.show()
```

- Andaman and Nicobar have Least number of discharged cases

### Active vs Deaths

### POPULATION AND TOTAL CASES

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='Total Cases', data=df)
plt.title('Total Cases vs. Population')
plt.xlabel('Population')
plt.ylabel('Total Cases')
plt.show()
```

```python
correlation_coefficient = df['Population'].corr(df['Total Cases'])
print(f"Correlation Coefficient between Population and Total Cases: {correlation_coefficient}")
```

- very weak negative correlation between Population and Total Cases
-  as the population of a state/union territory increases, the total number of COVID-19 cases might decrease
